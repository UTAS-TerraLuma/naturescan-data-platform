# 02 Create Thumbnails

```python
from ns_configs import RGB_ASSET_CONFIG, MS_ASSET_CONFIG
from ns_cogger import create_thumbnail
from pathlib import Path

force = True
limit = None
input_dir = Path("/Volumes/Sammy/terra-luma/stac-input")
```

## RGB Thumbnails

```python
# Get All RGB Tif Files
rgb_cogs = list(input_dir.rglob("*rgb.cog.tif", case_sensitive=False))

# Limit total (does nothing if limit is None)
rgb_cogs = rgb_cogs[0:limit]
print(f"Found {len(rgb_cogs)} RGB COGs.")

# Create a thumbnail for each COG
for i, input in enumerate(rgb_cogs):
    print(f"Processing {input.name} ({i + 1} / {len(rgb_cogs)}) ...")
    try:
        output = input.with_name(input.name.replace('.cog.tif', '.thumbnail.png'))
        create_thumbnail(input, output, force=force)
    except Exception as e:
        print(f"Error processing: {input.name}")
        print(e)
```
